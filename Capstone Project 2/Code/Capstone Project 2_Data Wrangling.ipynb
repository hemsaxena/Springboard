## Duplicate Data Checking

Looking for duplicate rows in the dataset now

```python
#Checking duplicate rows and storing in new dataframe duplicateRowsDF
duplicateRowsDF = df[df.duplicated()]
duplicateRowsDF
```

There are 1081 rows with duplicate values. It is worth assuming that same transaction at the same time, using the same amount is probably a data redundancy issue while recording the data. 

Checking if any of these duplicate rows have fraudulent transactions

```python
print("Number of duplicate fraudulent transactions are %s" %len(duplicateRowsDF[duplicateRowsDF['Class']==1]))
```

There are 19 rows pointing to duplicate fraudulent data as well.

This number of duplicate fraulent transaction data is not too high and therefore dropping them might not impact the dataset much, however I will investigate the impact further. 

So in the next step I will drop all duplicate rows including fraudulent as well as non-fraudulent row, but keep the 1st occurrence of each of these rows to maintain integrity of all unique occurrences

```python
#Dropping duplicate rows from the dataframe and storing into new dataframe dfn

dfn = df.drop_duplicates(keep='first')
print("The shape of the new dataset is now %s" %(str(dfn.shape)))
```

#### One final step - Checking impact of dropping duplicates

As per the metadata, there are 492 frauds out of 284,807 transactions which is only 0.172% of all transactions. So it is important to see the change in this percentage after dropping the duplicate rows

```python
percent_R = 100* (len(dfn[dfn['Class']==1])/len(dfn[dfn['Class']==0]))
print("\nIn the reduced dataset there are now %s%% frauds out of all transactions\n" %(np.round(percent_R,3)))
percent_O = 100* (len(df[df['Class']==1])/len(df[df['Class']==0]))

#Putting the shape of original and reduced data into a dataframe for visual comparison

dict ={'Dataset' : ['Original','Reduced'],'Fraud': [len(df[df['Class']==1]), len(dfn[dfn['Class']==1])],'Non-fraud':  [len(df[df['Class']==0]), len(dfn[dfn['Class']==0])],'Total':  [len(df), len(dfn)],'% of Frauds' :[str(np.round(percent_R,3)) + '%',str(np.round(percent_O,3)) + '%'] }

df_final = pd.DataFrame(dict)
df_final
```

The dataset size now has reduced by almost 0.006%.

Since the 'V' columns are PCA transformed, these repeated transactions might be associated with some valid transactions in the original dataset before PCA tranformation. So, dropping these might make the data inconsistent with the original data.

It could be useful to perform analysis with both datasets separately to see if there is any significant difference in the results when analysing the data by keeping all duplicates vs. dropping duplicates. So, I will maintain the data in both the states and try to do further analysis with both separately.
