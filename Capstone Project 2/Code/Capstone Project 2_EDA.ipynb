## Anamolies & Outliers - Review boxplots

```python
boxplot = df.boxplot(grid=False, vert=False,fontsize=15, figsize=(15, 15))
```

'Elapsed Time' has an extremely wider range of values as compared to other features.

'Amount' column seems to have some outliers, so I would make a boxplot of this column individually to investigate further as well.

```python
df.boxplot(column=['Amount'],return_type='axes')
```

There are many outliers which can be investigated further using interquantile range. Let's investigate that further

```python
Q1 = df.quantile(0.25)

# Now get Q3 and store in a variable called Q3. 
Q3 = df.quantile(0.75)

# Now calculate the IQR, storing it in a variable called IQR.
IQR = Q3- Q1

# Make a variable called `dfno`, and assign it the value: df[~((df < (Q1 - 1.5 * IQR)) |(df> (Q3 + 1.5 * IQR))).any(axis=1)]. 
# This filters on our existing dataframe, picking out just those observations that are NOT outliers. 
dfno = df[~((df < (Q1 - 1.5 * IQR)) |(df> (Q3 + 1.5 * IQR))).any(axis=1)]

# We now want to make a boxplot of this new dataframe dfno. 
boxplot = dfno.boxplot(grid=False, vert=False,fontsize=15, figsize=(12,15))
```

```python
#Comparing the shapes of the dataframes
dict1 = {"Dataframe" : ["Original","Reduced"], "Shape" : [df.shape,dfno.shape], "Fraudulent Transactions":[len(df[df['Class']==1]),len(dfno[dfno['Class']==1])]}
dispdf = pd.DataFrame(dict1)
dispdf
```

Removing outliers reduces the size of dataset significantly, but more importantly it clears out the entire fraudulent transaction data from the dataset. So outlier removal in this case is a bad idea.

# Investigating feature correlation

```python
sns.scatterplot(x='Amount',y='V2',data=df)
```

```python
sns.regplot(x='Amount',y='V2',data=df)
```

The line of best fit does not match the overall shape of the data, most of the points are clustered around one part of the line  rather than spreading evenly around the line. There doesn't seem very strong correlation by looking at the scatter plot and the line of best fit either, so I would go ahead with both the columns.

```python
from sklearn.metrics import mean_squared_error, r2_score # mean square error of regression and R^2 (coefficient of determination) regression score function.
from sklearn.model_selection import train_test_split #  Split arrays or matrices into random train and test subsets.
from sklearn import linear_model, preprocessing # Ordinary least squares Linear Regression.

X = df['V2'].values.reshape(-1,1)
Y = df['Amount'].values.reshape(-1,1)

# Split the data. This line uses the sklearn function train_test_split().
# The test_size parameter means we can train with 75% of the data, and test on 25%. 
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=123)
rModel = linear_model.LinearRegression()
rModel = rModel.fit(X_train,y_train)
rModel.score(X_train,y_train)
y_pred = rModel.predict(X_test)
plt.scatter(x=y_test,y=y_pred)
plt.xlabel('Actual Amount')
plt.ylabel('Predicted Amount')
print(rModel.score(X_train,y_train))
```

The score is quite low indicating that 'Amount' is unable to predict the values of 'V2' very good. Additionally, I'll apply the Ordinary Least Squares Algorithm to check if changing the algorithm can improve the R Squared value and if the model predicts the value better.

```python
import statsmodels.api as sm
X = df['V2'].values.reshape(-1,1)
Y = df['Amount'].values.reshape(-1,1)
X = sm.add_constant(X)
# Split the data. This line uses the sklearn function train_test_split().
# The test_size parameter means we can train with 75% of the data, and test on 25%. 
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=123)

rModel3 = sm.OLS(Y,X)

# Fit the model
rModel3 = rModel3.fit()
rModel3.summary()
```

The R Squared value has increased but still is low enough to conclude that 'Amount' does not give a good prediction of the values of 'V2'. So, we can proceed with both the columns('Amount' and 'V2') for further analysis.
